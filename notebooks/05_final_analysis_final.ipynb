# 5. Interpretación de métricas

La evaluación de modelos no se realizó utilizando únicamente accuracy.

Esto es importante porque:
- accuracy puede ser engañosa en problemas desbalanceados,
- y no refleja completamente la capacidad de detectar abandono.

Por esta razón se utilizaron múltiples métricas:

## Accuracy
Mide porcentaje total de aciertos.

## Precision
Mide cuántos positivos predichos realmente eran positivos.

## Recall
Mide cuántos clientes que realmente abandonan fueron detectados.

## F1-score
Equilibra precision y recall.

## ROC-AUC
Mide capacidad general de separación entre clases.

En este problema:
- recall y F1-score fueron especialmente importantes,
porque detectar clientes con riesgo de abandono tiene mayor valor estratégico para el negocio.
